Generating different types of toy data

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets.samples_generator import make_blobs
X, labels = make_blobs(n_samples=50, centers=10, n_features=7, random_state=5)
print(X.shape)

ytdist = X
Z = hierarchy.linkage(ytdist, 'single')
plt.figure()
dn = hierarchy.dendrogram(Z)
```

```python
from sklearn.cluster import DBSCAN
# Compute DBSCAN
db = DBSCAN(eps=5.0, metric='euclidean', min_samples=2).fit(X)
db.labels_
```

```python
import hdbscan
clusterer = hdbscan.HDBSCAN(min_cluster_size=3, gen_min_span_tree=True)
clusterer.fit(X)

clusterer.minimum_spanning_tree_.plot(edge_cmap='viridis',
                                      edge_alpha=0.6,
                                      node_size=80,
                                      edge_linewidth=2)

```

```python
clusterer.condensed_tree_.plot()
```

As we cxan see from the picture above, clustering with HDBSCAN works really well even if the cluster size is small. Such as 5 points.
